```python
import pandas as pd
```

```python
# Load the data
file_path = 'src/data/global-energy-substitution.csv'
df = pd.read_csv(file_path)
df
```

```python
# Combine the specified columns into a new 'Renewables' column
df['Renewables'] = df[['Other renewables', 'Biofuels', 'Solar', 'Wind', 'Hydropower', 'Nuclear']].sum(axis=1)

# Display first few rows of the modified data
df.head()
```

```python
# Save the modified data
modified_file_path = 'src/data/global-energy-substitution-modified.csv'
df.to_csv(modified_file_path, index=False)
```
